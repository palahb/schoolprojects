### Part 1.b)
The Cross-entropy Loss Function is calculated as $E_x(w) = - \sum_n [y^x_n \log{(w^x_n)} + (1-y^x_n) \log{(1-w^x_n)}]$, where: \
 $y^x_j = 1$ if $x$ is in class $j$ and $y^x_j = 0$ otherwise \
 $w^x_j$ is the probability of $x$ being in class $j$ \
 Then, deriving for each class we get $\frac{d E_x}{d w^x_j} = \frac{d}{d w^x_j} \sum_n [y^x_n \log{(w^x_n)} + (1-y^x_n) \log{(1-w^x_n)}] = \frac{d}{d w^x_j} y^x_j \log{(w^x_n)} - \frac{d}{d w^x_j} (1-y^x_j) \log{(1-w^x_n)} = \frac{y^x_j}{w^x_j} + \frac{1-y^x_j}{1-w^x_j}$

Before continuing with other parts, let us define a class for multinomial logistic regression:

#### Creating a Class for Multinomial Logistic Regression

In this phase, we will use Multinomial Logistic Regression classification model since we have more than two classes and cannot use Logistic Regression directly. So, we created a class for it to use later:

```python
class multLogReg(object):
    '''
    Class for multinomial logistic regression.
    '''

    def __init__(self):
        self.W = None # Weight
        self.b = None # Bias
    
    # Softmax function:
    def softmax(self, X):
        '''
        Softmax function
        '''
        # Linear mapping scores
        scores = np.dot(X, self.W)+self.b 

        # We normalized the scores beforehand with maximum value as zero
        # to avoid some problems with the exponential function:
        maxOfScores = np.max(scores, axis=1, keepdims=True) # Maximums
        expScores = np.exp(scores - maxOfScores) # Matrix of exponentials of scores

        # Sum of all exponentials:
        sumOfExps = np.sum(expScores, axis=1, keepdims=True)
        # Getting probabilities by softmax function:
        probabilities = expScores/sumOfExps

        return probabilities
    
    # Our loss function is cross-entropy loss:
    def crossEntropyLoss(self, X, y, reg_param, reg_type):
        '''
        Cross entropy loss with regularization terms
        '''

        nofSamples = X.shape[0]

        # Getting probabilities by softmax function:
        probs = self.softmax(X)

        # Logloss of the correct class for each of our samples
        correct_logprobs = -np.log(probs[np.arange(nofSamples), y])

        # Average loss:
        loss = np.sum(correct_logprobs)/nofSamples

        # Add regularization losses:
        if reg_type == 2: # If L2 regularization is used
            reg_loss = 0.5 * reg_param * np.linalg.norm(self.W,2) # 1/2 * \lambda |W|_2
        elif reg_type == 1: # If L1 regularization is used
            reg_loss = reg_param * np.linalg.norm(self.W,1) # \lambda |W|_1
        else: # If no regularization is used
            reg_loss = 0

        loss += reg_loss # Add regularization loss

        return probs, loss


    def celGradient(self, X, y, reg_param, reg_type):
        '''
        Gradient of cross entropy loss with regularization.
        '''
        nofSamples = X.shape[0]

        probs, _ = self.crossEntropyLoss(X, y, reg_param, reg_type)

        # Gradient of the loss with respect to scores
        dscores = probs.copy()

        # Substract 1 from the scores of the correct class
        dscores[np.arange(nofSamples),y] -= 1
        dscores /= nofSamples

        # Gradient of the loss with respect to weights
        dW = X.T.dot(dscores) 

        # Add gradient regularization: 
        if reg_type == 2: # L2 norm gradient
            dW += reg_param * self.W
            
        # Gradient of the loss with respect to biases
        db = np.sum(dscores, axis=0, keepdims=True)

        return dW, db


    def gradDesc(self, X, y, learning_rate=1e-4, reg_param=0, reg_type=0, nofIters=500):
        '''
        Gradent descent algorithm.
        '''
        nofFeatures, nofSamples = X.shape[1], X.shape[0]   
        nofClasses = len(np.unique(y))

        # Initialize weights and bias
        if (self.W is None) & (self.b is None):
            np.random.seed(2022) # for reproduciblity
            # Initialize weights from a random distribution
            self.W = np.random.normal(loc=0.0, scale=1e-4, size=(nofFeatures, nofClasses))
            # Initialize bias as zero
            self.b = np.zeros((1, nofClasses))

        for iter in range(nofIters):
            # Get gradients
            dW, db = self.celGradient(X, y, reg_param, reg_type)

            # update weights and biases
            self.W -= learning_rate*dW
            self.b -= learning_rate*db
            
            # If L1-regularization is used, we will do
            # soft thresholding:
            if reg_type == 1:
                for i in range(self.W.shape[0]):
                    for j in range(self.W.shape[1]):
                        # Soft thresholding:
                        if self.W[i][j] > reg_param:
                            self.W[i][j] -= reg_param
                        elif self.W[i][j] < -1 * reg_param:
                            self.W[i][j] += reg_param
                        else:
                            self.W[i][j] = 0

    def predict(self, X):
        prediction = np.argmax(np.dot(X, self.W)+self.b, axis=1)
        return prediction
       
```

Here, a function for calculating accuracies:

```python
def getAccuracy(prediction, actual):
    nofCorrectPredictions = 0
    for i in range(len(prediction)):
        if prediction[i] == actual[i]:
            nofCorrectPredictions += 1
    accuracy = (nofCorrectPredictions / len(prediction))
    return accuracy
```

### Part 1.c)
Here, a generic 5-fold cross validation function:

```python
def cross_validation(X, y, validation_type, params, LR = 0.01):
    '''
    5 fold cross validation for determining learning rate of the gradient descent algorithm
    or determining L1 or L2 regularization term.
    '''
    train_data = X.copy()
    train_label = y.copy()
    
    partitions_data = np.array_split(train_data, 5)
    partitions_label = np.array_split(train_label, 5)
    
    accuracies = {} # Accuracies for all parameters
    for param in params:
        
        acc = [] # Accuracies of a parameter for every run
        
        for i in range(5): # Cross validation
            train_data, train_label = [], []
            test_data, test_label = [], []
            
            for j in range(5): # Create test and train data and labels
                if i == j:
                    test_data.extend(partitions_data[j])
                    test_label.extend(partitions_label[j])
                else:
                    train_data.extend(partitions_data[j])
                    train_label.extend(partitions_label[j])
                    
            train_data, train_label = np.array(train_data), np.array(train_label)
            test_data, test_label = np.array(test_data), np.array(test_label)
            
            mult_log_reg = multLogReg()
            
            # In this project, we used cross validation for three purposes:
            if validation_type == "learning-rate": # To determine learning rate
                mult_log_reg.gradDesc(train_data, train_label, learning_rate = param)
                
            elif validation_type == "l2-regularization": # To determine L2-regularization parameter
                mult_log_reg.gradDesc(train_data, train_label, learning_rate = LR, reg_param = param, reg_type = 2)
                
            elif validation_type == "l1-regularization": # To determine L1-regularization parameter
                mult_log_reg.gradDesc(train_data, train_label, learning_rate = LR, reg_param = param, reg_type = 1)
                
            else:
                raise NotImplementedError
            
            acc.append(getAccuracy(mult_log_reg.predict(test_data), test_label))
        
        accuracy = np.array(acc)
        
        accuracies[param] = acc

    return accuracies
```

We considered 10 possible learning rates and cross validated them:

```python
learning_rates = [0.00001, 0.00005, 0.0001, 0.0005, 0.001, 0.005, 0.01, 0.05, 0.1, 0.5]
accuracies = cross_validation(X_train, y_train, validation_type = "learning-rate", params = learning_rates)
```

Here is the results:

```python
for rate, acc in accuracies.items():
    mean, std = np.mean(acc), np.std(acc)
    print(f"For learning rate {rate}, mean accuracy is {mean:.2%}, std is {std:4f}")
```

Best is achieved when the learning rate = 0.01

### Part 1.h)

When we use L1 regularization, our objective function $f$ becomes

$ f = E_{in} + \lambda |\textbf{w}|$

To implement L1-regularization, we did a research about soft thresholding and learned that to incorporate the L1 norm into gradient descent an operator called _soft thresholding operator_ $S_\alpha$ is used.

$ S_\alpha(x) = \begin{cases}  x + \alpha & \text{if } x \leq -\alpha \\ 0 & \text{if }|x| < \alpha \\ x - \alpha & \text{if } x \geq \alpha \end{cases}$

We implemented this soft thresholding in our gradient descent. We checked every entry $\textbf{w}_{ij}$ of our weight vector $\textbf{w}$ and performed soft thresholding:

$ S_\alpha(\textbf{w}_{ij}) = \begin{cases}  \textbf{w}_{ij} + \alpha & \text{if } \textbf{w}_{ij} \leq -\alpha \\ 0 & \text{if }|\textbf{w}_{ij}| < \alpha \\ \textbf{w}_{ij} - \alpha & \text{if } \textbf{w}_{ij} \geq \alpha \end{cases}$

You can refer to the class [here](#Creating-a-Class-for-Multinomial-Logistic-Regression).

```python
l1_reg_params = [1e-5, 2.5e-5, 5e-5, 1e-4, 2.5e-4, 5e-4, 0.001, 0.0025, 0.005, 0.01]
accuracies = cross_validation(X_train, y_train, validation_type = "l1-regularization", params = l1_reg_params)
```
